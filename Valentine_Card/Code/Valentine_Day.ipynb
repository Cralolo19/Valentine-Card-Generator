# Making the plot

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patheffects as pe

# Choosing the matplotlib style
plt.style.use('sty.mplstyle')

# Creating the fig and ax instances required to make the plot
fig = plt.figure(figsize=(13,13))
ax = fig.add_subplot(111, projection='polar')

# Choosing the facecolor for Fig and Ax instances
color = "#800020"
ax.set_facecolor(color)
fig.patch.set_facecolor(color)

# Generating the data needed to create a heart
theta = np.linspace(-np.pi/2, 3*np.pi/2, 5000)
r = 2 - 2*np.sin(theta) + np.sin(theta)*np.sqrt( np.abs(np.cos(theta)) )/( np.sin(theta) + 1.45)

# Plotting the heart shape
ax.plot(theta, r, color='white',path_effects=[pe.Stroke(linewidth=4, foreground='red'), pe.Normal()], zorder=20)
for i in range(1,11):
    j = i/10
    ax.plot(theta, (j-1/10)*r, color='white', lw=1, alpha=1, zorder=10)
    ax.plot(theta, (j-1/10)*r, color='pink', lw=4, alpha=1, zorder=0)
    ax.fill_between(theta, (j-1/10)*r, j*r, color='white', alpha=1, zorder=-100)
    ax.fill_between(theta, (j-1/10)*r, j*r, color='red', alpha=1.1-j, zorder=10-i)

# Adding messages to the Valentine's card
main_message = r"\underline{\it Happy Valentine's Day!}"
second_message = r"{\it From My Heart \\ To Yours}"

plt.gcf().text(0.5,0.78,main_message, color='white', alpha=1,fontsize=55, ha='center', path_effects=[pe.Stroke(linewidth=1, foreground='k'), pe.Normal()])
plt.gcf().text(0.5, 0.65, second_message, color='white', alpha=1,fontsize=55, ha='center', path_effects=[pe.Stroke(linewidth=1, foreground='k'), pe.Normal()])

# Setting the x and y axes
ax.spines['polar'].set_visible(False)
ax.xaxis.set_ticks([])
ax.yaxis.set_ticks([])
ax.set_ylim([0, 4])
ax.set_aspect('equal')

# Adding an arrow to the heart
ax.annotate('', xy=(0.85, 0.65), xycoords='axes fraction', xytext=(0.18,0.12),
            arrowprops=dict(arrowstyle='fancy', facecolor='#F88379', edgecolor='white', linewidth=2),
            clip_on=False, size=110, zorder=-200)

# Saving the Valentine's card
plt.savefig('../Plots/ValentineDay.pdf', transparent=False, bbox_inches='tight')
plt.savefig('../Plots/ValentineDay.png', transparent=False, bbox_inches='tight', dpi=500)
```
